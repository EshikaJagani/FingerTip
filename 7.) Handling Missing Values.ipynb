#  Filling null values with the previous ones

```python
data=pd.read_csv("Missing_Values.csv",parse_dates=["Date"])
data
```

```python
#Replace NaN with Forward fill
new_data = data.fillna(method="ffill") 
new_data
```

Filling null value with the next ones using **Backward fill** method

# Backward fill to carry forward next day's value

```python
#Replace NaN with Backward fill
new_data = data.fillna(method="bfill")
new_data
```

Using **Axis** to fill value horizontally i.e from the neighbouring cell

```python
data
```

```python
#Fill values by side of the columns (which in our case is not helpful)
new_data = data.fillna(method="bfill",axis="columns")
new_data
```

**Limit** is used to fill multiple consecutive NaN values by limiting the ffill/bfill from previous/next cell not to all, i.e. limit 1 upto 1 cell, limit 2 upto 2 cell

```python
data
```

```python
new_data = data.fillna(method="ffill",limit=1)
new_data
```

```python
new_data = data.fillna(method="ffill",limit=3)
new_data
```

# Keep the row if it has any single non NaN value, threshold=1, keep row if it has one valid value one if threshold = 2, it will keep if there are 2 valid values

```python
data.set_index('Date',inplace=True)
```

```python
data
```

```python
#Keep the rows if it has any non na value
#thershlod parameter(We need one valid value in order to keep the row)
#thresh=N requires that a column has at least N non-NaNs to survive
new_data=data.dropna(thresh=1)
new_data
```

```python
#We want 2 valid(non na) value in order to keep the row
new_data=data.dropna(thresh=2)
new_data
```
