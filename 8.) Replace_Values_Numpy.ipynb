```python
#Replacing list with single value
new_data = data.replace(['32 Years','34'],0)
new_data
```

```python
#Replacing per column
new_data = data.replace({
        'Age': '32 Years',
        'Sales': '6 Cr',
        'Day': 'Tuesday'
    }, {
        'Age': '34 Years',
        'Sales': '10 Cr',
        'Day': 'No Day'
    })
new_data
```

```python
#Replacing special value by using mapping


new_data = data.replace({
        '32 Years': np.nan,   
        np.nan: 'Sunday',
    })
new_data
```

```python
#Regex-Regular Expression
# when Age is 32 Years, Sales is 6 lacs etc.
new_data = data.replace({
    'Age': '[A-Za-z]', 
    'Sales': '[A-Za-z]'
},'', regex=True) 
new_data
```

```python
#Replacing list with another list
#creating a new data frame
df = pd.DataFrame({
    'score': ['exceptional','average', 'good', 'poor', 'average', 'exceptional'],
    'student': ['Maya', 'Rakesh', 'Rajiv', 'Harry', 'jinal', 'Shubham']
})
df
```

```python
new_df=df.replace(['poor', 'average', 'good', 'exceptional'], [1,2,3,4])
new_df
```
